```python
from sklearn.datasets import load_iris
import pandas as pd
import matplotlib.pyplot as pp
```

```python
iris=load_iris()
```

```python
df=pd.DataFrame(data=iris.data,columns=iris['feature_names'])
```

```python
df.head()
```

```python
df.shape
```

```python
df['target']=iris.target
```

```python
df.target.value_counts()
```

```python
import seaborn as sns
sns.countplot(x=df.target,data=df)
```

```python
ax1,ax2,ax3,ax4 = pp.figure(figsize = (16,5)).subplots(1,4)
sns.lineplot(data=df, x=df['sepal length (cm)'], y=df['target'], ax=ax1)
sns.lineplot(data=df, x=df['sepal width (cm)'],y=df['target'], ax=ax2)
sns.lineplot(data=df, x=df['petal length (cm)'], y=df['target'], ax=ax3)
sns.lineplot(data=df, x=df['petal width (cm)'], y=df['target'], ax=ax4)

pp.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
x=df.drop(['target'],axis=1)
y=df.target
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.1,random_state=1)
```

```python
model=LogisticRegression()
model.fit(xtrain,ytrain)
```

```python
ypre=model.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(ypre,ytest)
```
