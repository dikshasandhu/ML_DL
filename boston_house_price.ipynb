```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pp
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_absolute_error,mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
```

```python
columns = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV']
df=pd.read_csv('boston_house.csv',header=None,delimiter=r"\s+",names=columns)
```

```python
x=df.drop(['MEDV'],axis=1)
y=df.MEDV
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.5,random_state=10)
```

```python
models = [LinearRegression(), 
          DecisionTreeRegressor(), 
          RandomForestRegressor(),  
          GradientBoostingRegressor()
         ]

model_names = ['Linear Regression', 
               'Decision Tree', 
               'Random Forest',  
               'Gradient Boosting Regression'
              ]
```

```python
performances = {}
for model, name in zip(models, model_names):
    model.fit(xtrain, ytrain)
    ypred = model.predict(xtest)
    mae = mean_absolute_error(ytest, ypred)
    mse = mean_squared_error(ytest, ypred)
    print('mae score for model-{}:{}'.format(name,mae))
    print('mse score for model-{}:{}'.format(name,mse))
```
